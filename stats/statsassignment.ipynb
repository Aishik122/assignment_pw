Q= Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
A>ter generating the list, >ind the >ollowing:

```python
import random ,math
int_list =[random.randint(90,130) for x in range(100)]
```

```python
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        return (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2
    else:
        return sorted_numbers[n//2]

def calculate_mode(numbers):
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1
    max_frequency = max(frequency.values())
    modes = [num for num, freq in frequency.items() if freq == max_frequency]
    return modes
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight
def calculate_geometric_mean(numbers):
    product = 1
    for num in numbers:
        product *= num
    return math.pow(product, 1 / len(numbers))
def calculate_harmonic_mean(numbers):
    reciprocal_sum = sum(1 / num for num in numbers)
    return len(numbers) / reciprocal_sum
def calculate_midrange(numbers):
    return (max(numbers) + min(numbers)) / 2
def calculate_trimmed_mean(numbers, percentage):
    sorted_numbers = sorted(numbers)
    trim_length = int(len(sorted_numbers) * (percentage / 100) / 2)
    trimmed_numbers = sorted_numbers[trim_length:-trim_length]
    return sum(trimmed_numbers) / len(trimmed_numbers)
```

```python
print(f"calculate_mean is : {calculate_mean(int_list)}")
print(f"calculate_median is : {calculate_median(int_list)}")
print(f"calculate_mode is : {calculate_mode(int_list)}")
# print(f"calculate_weighted_mean is : {calculate_weighted_mean(int_list,[x for x in range 100])}") 
print(f"calculate_geometric_mean is : {calculate_geometric_mean(int_list)}")
print(f"calculate_harmonic_mean is : {calculate_harmonic_mean(int_list)}")
print(f"calculate_midrange is : {calculate_midrange(int_list)}")
print(f"calculate_trimmed_mean is : {calculate_trimmed_mean(int_list,20)}")
```

 2. Generate a list o> 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
A>ter generating the list, >ind the >ollowing:

```python
import statistics
```

```python
int_list2 =[random.randint(200,300) for x in range(500)]
```

```python
def calculate_range(numbers):
    return max(numbers) - min(numbers)

def calculate_variance_and_stddev(numbers):
    variance = statistics.variance(numbers)
    stddev = statistics.stdev(numbers)
    return variance, stddev
def calculate_iqr(numbers):
    numbers.sort()
    q1_index = (len(numbers) - 1) // 4
    q3_index = q1_index * 3
    q1 = numbers[q1_index]
    q3 = numbers[q3_index]
    return q3 - q1
    
def calculate_coefficient_of_variation(numbers):
    mean = statistics.mean(numbers)
    stddev = statistics.stdev(numbers)
    return (stddev / mean) * 100 if mean != 0 else 0
def calculate_mad(numbers):
    mean = statistics.mean(numbers)
    mad = statistics.mean(abs(num - mean) for num in numbers)
    return mad
def calculate_quartile_deviation(numbers):
    if len(numbers) < 4:
        return None  
    numbers.sort()
    q1_index = (len(numbers) - 1) // 4
    q3_index = q1_index * 3
    q1 = numbers[q1_index]
    q3 = numbers[q3_index]
    return (q3 - q1) / 2
def calculate_range_based_cod(numbers):
    return (max(numbers) - min(numbers)) / (max(numbers) + min(numbers))

```

```python
print(f"calculate_range is : {calculate_range(int_list2)}")
print(f"calculate_variance_and_stddev is : {calculate_variance_and_stddev(int_list2)}")
print(f"calculate_iqr is : {calculate_iqr(int_list2)}")
print(f"calculate_coefficient_of_variation is : {calculate_coefficient_of_variation(int_list2)}") 
print(f"calculate_mad is : {calculate_mad(int_list2)}")
print(f"calculate_quartile_deviation is : {calculate_quartile_deviation(int_list2)}")
print(f"calculate_range_based_cod is : {calculate_range_based_cod(int_list2)}")
```

)  Write a Pthon class representing a discrete random variable with methods to calculate its expected
value and variance.

```python
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        if len(outcomes) != len(probabilities):
            raise ValueError("Number of outcomes and probabilities must be equal.")
        
        self.outcomes = outcomes
        self.probabilities = probabilities
    
    def expected_value(self):
        """
        Calculate the expected value of the discrete random variable.
        
        Returns:
        float: Expected value.
        """
        return sum(outcome * probability for outcome, probability in zip(self.outcomes, self.probabilities))
    
    def variance(self):
        """
        Calculate the variance of the discrete random variable.
        
        Returns:
        float: Variance.
        """
        expected_value = self.expected_value()
        return sum(probability * (outcome - expected_value) ** 2 for outcome, probability in zip(self.outcomes, self.probabilities))
```

0) Implement a program to simulate the rolling o6 a 6air six-sided die and calculate the expected value and
variance o6 the outcomes.

```python
import random

class DiceSimulation:
    def __init__(self, sides=6, rolls=1000):
        self.sides = sides
        self.rolls = rolls
    
    def simulate(self):
        outcomes = [random.randint(1, self.sides) for _ in range(self.rolls)]
        return outcomes
    
    def expected_value(self):
        return sum(range(1, self.sides + 1)) / self.sides
    
    def variance(self):
        return sum((i - self.expected_value()) ** 2 for i in range(1, self.sides + 1)) / self.sides

dice_sim = DiceSimulation()

outcomes = dice_sim.simulate()

expected_value = dice_sim.expected_value()
print("Expected Value:", expected_value)

variance = dice_sim.variance()
print("Variance:", variance)
```
