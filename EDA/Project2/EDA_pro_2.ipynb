### Data Cleaning

```python
df['category'].unique()
```

after observing above unique product category we can say first substring denote product category and last denote product type so we can extract only this prodct and drop category because it didnt make any sence 

```python
df['product category']=df['category'].apply(lambda x:x.split('|')[0])
```

```python
df['product category'].unique()
```

```python
df['product_type']=df['category'].apply(lambda x:x.split('|')[-1])
```

```python
df['product_type'].unique()
```

```python
df['product_type'].nunique()
```

```python
df
```

```python
df.drop('category',inplace=True,axis=1)
```

```python
df.sample()
```

```python
df['discounted_price'] = df['discounted_price'].apply(lambda x: float(x.replace('₹', '').replace(',', '')))
```

```python
df['actual_price']=df['actual_price'].apply(lambda x: float(x.replace('₹', '').replace(',', '')))
```

```python
df['discount_percentage']=df['discount_percentage'].apply(lambda x: float(x.replace('%', '').replace(',', '')))
```

```python
df['rating']=df['rating'].apply(lambda x: float(x.replace('|', '0')))
```

```python
df['rating_count']=df['rating_count'].apply(lambda x: float(str(x).replace(',', '')))
```

```python
df['rating_count'].fillna(0,inplace=True)
```

1. What is the average rating for each product category

```python
category_rating=pd.DataFrame(df.groupby('product category')['rating'].mean())
```

```python
category_r_fig = go.Figure()
category_r_fig.add_trace(go.Bar(x=category_rating.index,y=category_rating.rating,marker_color='skyblue'))
category_r_fig.update_layout(title = "Category wise Rating average",xaxis_title = 'Categories',yaxis_title='Values')
```

2. What are the top rating_count products by category

```python
prod_cate=df.groupby(['product category','product_type'],as_index=False)['rating_count'].sum()
prod_cate['rank'] = prod_cate.groupby('product category')['rating_count'].rank(method='dense', ascending=False)
prod_cate=prod_cate[prod_cate['rank']==1]
prod_cate
```

```python
category_rc_fig = go.Figure()
category_rc_fig.add_trace(go.Bar(x=prod_cate['product_type'],y=prod_cate.rating_count,marker_color='skyblue'))
category_rc_fig.update_layout(title = "Category wise Rating count average",xaxis_title = 'Categories',yaxis_title='Values')
```

6. What are the most popular product keywords?

```python
# keywords  are generally present in the about product description.
s={}
for i in df['product_name']:
    x=i.split()
    for j in x:
        if j not in s.keys():
            s[j]=1
        else:
            s[j]+=1
print(s)
```

```python
sorted_dict = dict(sorted(s.items(), key=lambda item: item[1], reverse=True))
print(sorted_dict)
```

most frequent key words are -> usb,Cable,Charging,Bluetooth,Black,Remote,Heater
